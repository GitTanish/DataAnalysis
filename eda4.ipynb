####Flight Price Prediction(EDA + Feature Engineering)

```python
#importing basic libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df=pd.read_excel('Data_Train.xlsx')
train_df.head()
```

```python
test_df=pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
final_df = pd.concat([train_df, test_df], ignore_index=True)
final_df.head()
```

```python
final_df.tail()
```

```python
final_df.info()
```

```python
# firstly, date of journey is in object form, and break it into date, month and year, basically a deriverd feature
```

```python
### Feature Engineering process
final_df['Date_of_Journey'] = pd.to_datetime(final_df['Date_of_Journey'])
final_df['Date'] = final_df['Date_of_Journey'].dt.day
final_df['Month'] = final_df['Date_of_Journey'].dt.month
final_df['Year'] = final_df['Date_of_Journey'].dt.year
```

```python
final_df.head(2)
```

```python
final_df.drop("Date_of_Journey",axis=1, inplace = True)
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min']=final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head()
```

```python
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)
```

```python
final_df.drop('Arrival_Time', axis=1,inplace=True)
```

```python
final_df['Dept_hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dept_min']=final_df['Dep_Time'].str.split(':').str[1]
final_df['Dept_hour']=final_df['Dept_hour'].astype(int)
final_df['Dept_min']=final_df['Dept_min'].astype(int)
final_df.drop('Dep_Time', axis=1, inplace=True)
```

```python
final_df.info()
```

```python
final_df['duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df[final_df['duration_hour']=='5m']
# i think i should drop the 5m value as it looks incorrect logically
# how can mumbai to hyderabad take 5 min only
final_df = final_df[final_df['duration_hour'] != '5m']
```

```python
final_df['duration_hour']=final_df['duration_hour'].astype('int')
```

```python
final_df['duration_hour']*60
```

```python
final_df['duration_min'] = final_df['Duration'].str.split(' ').str[1].str.split('m').str[0].fillna(0)
final_df['duration_min']= final_df['duration_min'].astype(int)
```

```python
final_df['duration_min'] = final_df['duration_hour']*60 + final_df['duration_min']
```

```python
final_df.head()
```

```python
final_df.drop('Duration', axis = 1, inplace=True)
```

```python
final_df.info()
```

```python
final_df.drop('duration_hour', axis = 1, inplace=True)
```

```python
final_df.drop('Route', axis = 1, inplace=True)
```

```python
final_df.info()
```

```python
final_df['Airline'].unique()
```

```python
# label encoding for the Airline and rest of categorical features
from sklearn.preprocessing import LabelEncoder
labelencoder= LabelEncoder()
```

```python
final_df['Airline']=labelencoder.fit_transform(final_df['Airline'])
final_df['Source']=labelencoder.fit_transform(final_df['Source'])
final_df['Destination']=labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']=labelencoder.fit_transform(final_df['Additional_Info'])
```
