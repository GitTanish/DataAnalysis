####1. Import the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/air-passengers/AirPassengers.csv')
```

####2. Describe the dataset

```python
# 2. Describe the dataset
print("\n--- Dataset Description ---")

    # i. Structure
print(f"Shape (Rows, Columns): {df.shape}")

    # ii. Data types
print("\nData Types:\n", df.dtypes)

    # iii. Missing values
print("\nMissing Values (Count per column):\n", df.isnull().sum())

    # iv. Range of dates
df['Month'] = pd.to_datetime(df['Month'])
print("\nRange of Dates:")
print(f"Start Date: {df['Month'].min()}")
print(f"End Date: {df['Month'].max()}")
```

#### 3.Overall trends

```python
print("\n--- Overall Trend in Number of Passengers ---")
plt.plot(df['Month'], df['#Passengers'])
plt.title('Overall Trend in Air Passengers')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()
```

####4. Outliers



```python
print("\n--- Outliers ---")
plt.figure(figsize=(8, 5))
sns.boxplot(y=df['#Passengers'])
plt.title('Box Plot of Number of Passengers')
plt.ylabel('Number of Passengers')
plt.show()
print("Outliers can be visually inspected in the box plot. Points outside the whiskers are potential outliers.")
```
