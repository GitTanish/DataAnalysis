```python
## let's see which category has most no. of mean subscribers
```

```python
pd.set_option("display.float_format", "{:.2f}".format)
```

```python
category_subs_mean = df.groupby("category")["subscribers"].mean()
category_subs_mean.sort_values(ascending=False)
```

```python
category_subs_mean.plot(kind='bar')
plt.title("Mean Subscribers of category")
plt.xlabel("Category")
plt.ylabel("mean Subscribers")
plt.show()
```
